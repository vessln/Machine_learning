## 3. Merging the two datasets

Make the values ​​in the corresponding columns of both datasets be in the same format:

```python
dataset2['gender'] = dataset2['gender'].apply(lambda x: 'M' if x == 1 else 'F')
```

```python
dataset2.gender.unique()
```

```python
map_types = {0: 'TA', 1: 'TA', 2: 'ATA', 3: 'NAP', 4: 'ASY'}
dataset2['chest_pain_type'] = dataset2['chest_pain_type'].map(map_types)
```

```python
dataset2
```

```python
dataset2.exercise_angina.unique()
```

```python
dataset2['exercise_angina'] = dataset2['exercise_angina'].apply(lambda x: 'Y' if x == 1 else 'N')
```

```python
map_types_ecg = {0: 'Normal', 1: 'ST', 2: 'LVH'}
dataset2['resting_ecg'] = dataset2['resting_ecg'].map(map_types_ecg)
```

I assume that the values ​​of st_slope 0 and 1 in dataset2 indicate that the ST segment slopes UPward and usually show normal heart function, so i merge them.

```python
map_slope_types = {0: 'Up', 1: 'Up', 2: 'Flat', 3: 'Down'}
dataset2['st_slope'] = dataset2['st_slope'].map(map_slope_types)
```

```python
merged_dataset = pd.concat([dataset1, dataset2], ignore_index = True)
```
