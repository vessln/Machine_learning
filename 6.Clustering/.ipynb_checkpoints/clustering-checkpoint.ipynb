# Clustering

## Clustering analysis

Clustering analysis is a method of grouping data into clusters so that objects in one cluster are more similar to each other than to objects from other clusters. The goal is to divide the data into clusters based on some measure of similarity (most commonly distance). Clustering is based on features or graphs. This is a type of unsupervised learning because there is no target or predefined labels for the model to predict. 

### Application
- Grouping customers based on their purchases (Marketing Sales);
- Image analysis – clustering of pixels for segmentation (Object recognition, Computer vision);
- Genomic data clustering – clustering of genes with similar functions;
- Anomaly detection (Fraud detection, Network security);

## K-Means Clustering

**K-Means Clustering** sepatares the data into *k* number of clusters. The idea is to group objects so that points in one cluster are closer to the "mean" point of that cluster than to the midpoint of another cluster. Steps:
1. I set the number of clusters k – how many groups I want to have;
2. I choose random points (seeds) for centers of these clusters (centroids);
3. I allocate each point to the nearest centroid (by distance);
4. I calculate the new centroids - these are new points, each of which is equal to the average value of all points in the cluster (in this step, some points may go from one to another cluster because they are closer to the new centroid); This is repeated until the centroids stop changing significantly; \
The algorithm seeks to minimize the distance of each point to its centroid by constantly reordering the points and changing the clusters.
K-Means clustering is the simplest clustering algorithm and is very fast.

### Implementation

```python
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2) ** 2))


def kmeans(X, k, max_iters = 100):
    np.random.seed(42)
    
    # 1. Randomly initialize centroids
    centroids = X[np.random.choice(X.shape[0], k, replace=False)]
    print(f"Initial centroids:\n{centroids}\n")
    
    for i in range(max_iters):
        print(f"Iteration {i+1}:")
        
        # 2. Assign each point to the nearest centroid
        clusters = [[] for _ in range(k)]
        labels = []
        for point in X:
            distances = [euclidean_distance(point, centroid) for centroid in centroids]
            closest_centroid = np.argmin(distances)
            clusters[closest_centroid].append(point)
            labels.append(closest_centroid)
        
        # 3. Calculate new centroids as the mean of assigned points
        new_centroids = np.array([np.mean(cl, axis=0) if cl else centroids[i] for i, cl in enumerate(clusters)])
        print(f"New centroids:\n{new_centroids}\n")
        
        # If centroids do not change, we have converged
        if np.all(centroids == new_centroids):
            print("Convergence reached.")
            break
        
        centroids = new_centroids

    colors = ['r', 'g', 'b']
    for j, cluster in enumerate(clusters):
        cluster_points = np.array(cluster)
        if len(cluster_points) > 0:
            plt.scatter(cluster_points[:, 0], cluster_points[:, 1], color=colors[j])
    
    plt.scatter(centroids[:, 0], centroids[:, 1], color='k', marker='x', s=150, label='Centroids')
    plt.title("Clusters and Centroids")
    plt.xlabel("feature 1")
    plt.ylabel("feature 2")
    plt.show()
    
    return labels
```

```python
X = np.array([[6, 7], [1, 3], [2.2, 4], [1.8, 3.1], [6, 7.9], [11, 13], [7.2, 10.2], [7, 8], [3.3, 1.7], [13, 11.6]])
kmeans(X, 3)
```

**Parameters:**
- n_clusters – number of clusters *k*;
- max_iter – the maximum number of iterations to update the centroids. If the model does not converge, it stops when it reaches value of max_iter;
- tol – when the change in centroids between two iterations is less than this tolerance, the algorithm assumes it has converged and stops;
- init – set the method for initializing the initial centroids:
    - random – all centroids are chosen randomly.
    - k-means++ (by default) – the first centroid is randomly selected, and each subsequent one is selected so that it is far from the already selected ones. The probability of selecting a new centroid that is close to an already selected one is very small.
    - I can set initial centroids with np.array.

### Cluster inertia (loss function) 

Cluster inertia is K-Means loss function. It finds every distance from one point to every centroid and chooses the smallest one. The sum of all these distances for all points gives the total loss function.

Аs the number of centroids increases, the inertia decreases. The smaller the inertia, the data is better clustered. 

```python
# the value of loss function:
kmeans3.inertia_
```

How to choose the number of centroids *k* ? 
- Ewbol method (WCSS) is a method for finding the number of optimal centroids by choosing the point where there is an elbow / where the momentum stops decreasing significantly. If there is no distinct bend, then there is not an appropriate number of clusters.
- Silhouette analysis – the ratio between the distance between a point and its nearest centroid AND the distance between this point and its second nearest centroid. Silhouette Score is close to 1 (perfect) if the point is much closer to its cluster than to the others. Silhouette analysis helps determine the optimal number of clusters. With this analysis, the quality of the clustering can be evaluated - the ratio between distance of each point and its nearest centroid AND the average distance between clusters. The higher it is, the better the clustering.

```python
data = {"n_clusters": [], "inertia_values": []}

for k in range(1, 11):
    kmeans_test = KMeans(n_clusters = k)
    kmeans_test.fit(attributes)
    data["n_clusters"].append(k)
    data["inertia_values"].append(kmeans_test.inertia_)
```

```python
n_clusters = 3
kmeans_slt = KMeans(n_clusters = n_clusters, random_state = 42)
kmeans_slt.fit(attributes)

plt.scatter(attributes[:, 0], attributes[:, 1], c = kmeans_slt.predict(attributes))
plt.scatter(kmeans_slt.cluster_centers_[:, 0], kmeans_slt.cluster_centers_[:, 1], c = "black", marker='x')
plt.show()
```

```python
labels = kmeans_slt.predict(attributes)

silhouette_vals = silhouette_samples(attributes, labels)
avg_silhouette = silhouette_score(attributes, labels)

print(f"Average silhouette score = {avg_silhouette:.3f}")

plt.figure(figsize = (6, 5))

y_lower = 10
for i in range(n_clusters):
    ith_cluster_silhouette_vals = silhouette_vals[labels == i]
    ith_cluster_silhouette_vals.sort()
    
    size_cluster_i = ith_cluster_silhouette_vals.shape[0]
    y_upper = y_lower + size_cluster_i
    
    color = plt.cm.nipy_spectral(float(i) / n_clusters)
    plt.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_silhouette_vals, facecolor=color, edgecolor=color, alpha=0.7)
    
    plt.text(-0.03, y_lower + 0.5 * size_cluster_i, str(i + 1))
    y_lower = y_upper + 10


plt.axvline(x = avg_silhouette, color = "red", linestyle = "--")
 
plt.title("Silhouette plot")
plt.xlabel("silhouette coefficient values")
plt.ylabel("cluster")

plt.show()
```

With kmeans.fit(attributes) the model is trained on the attributes. In the case of K-Means, this means that the centroids of the clusters are calculated, the goal being to find the best centroids that minimize the distance from each point to its nearest centroid. This process uses a loss function that measures the sum of the squares of the distances between points and their centroids (inertia). \
With kmeans.predict(attributes) the model uses the centroids found by .fit() and for each point in attributes predicts which cluster it belongs to. This is done by calculating the distance between each point and all centroids, assigning the point to the closest cluster.

## Hierarchical clustering 

Hierarchical clustering – a data grouping method that finds a hierarchy from which it creates clusters. A hierarchy represents an asymmetric relationship.

### Agglomerative Clustering
Agglomerative Clustering is a method to find a hierarchy. It is not used for prediction! \
Steps: it starts from the smallest units (individual points) and gradually merges into larger clusters: Initially, each point is a separate cluster. And at each step, two clusters (points) with the smallest distance between them are merged into a new larger cluster. This merging continues until there is only one large cluster left that contains all the points. After all the points are merged into one cluster, I can tell at which stage to stop the merging and choose the final number of clusters.

## DBSCAN

DBSCAN (Density-based Spatial Clustering of Applications with Noise) is used for data with outliers. If there is no outliers in the dataset, it's better to use another clustering algorithm. There is no need to give the number of clusters in advance. DBSCAN finds the clusters by itself, it also finds those of arbitrary shape. Parameters:
- eps – the radius of each point where the model searches for neighbors. Too small a value will result in very small clusters, and too large will merge too many different clusters into one.
- min_samples – the minimum number of points a point must have within eps to be considered key. It is often chosen based on the number of features.
  
Algorithm steps: Picks a random point. Checks if the point has at least min_samples neighboring points within radius eps. If yes, it becomes a key point and forms a new cluster. All neighboring points are added to the cluster, incl. their neighbors if they are also key points. This continues until all points in the cluster are found. If a point is not key and cannot form a cluster, it is considered noise.
