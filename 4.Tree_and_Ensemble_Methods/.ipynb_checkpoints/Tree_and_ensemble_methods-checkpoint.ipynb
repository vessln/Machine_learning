The tree is very large, so I want to reduce the number of questions it asks, as regularization in linear models. To make the model simpler I can adjust some of its parameters (max_depth / min_samples_split / min_samples_leaf).

Hyperparameters are set separately:
- criterion – defines the function to measure the separation quality. By default it is gini (more often used), but it can also be entropy (with a lot of data).
- splitter – defines the splitting selection strategy. By default = best (selects the best possible split); random
- max_depth – defines the maximum depth of the tree. If the value is too high, the model may be overfitting. Shallow depth results in a simpler model that may be more resistant to overfitting.
- min_samples_split – minimum number of records needed to split a node. A high value limits branching and reduces the risk of overfitting.
- min_impurity_decrease – the minimum impurity decrease for a node to be split. This is an impurity threshold that ensures that a separation will only occur if it results in a significant reduction of the impurity.
- min_samples_leaf – minimum number of records each leaf should have. A high value creates a simpler tree and reduces the risk of overfitting.
- max_features – the number of features to consider in each split. By default = None – all are considered. Reducing the number of features can help reduce model complexity.
- max_leaf_nodes – limits the number of leaves.
- class_weight – weights for the different classes to compensate for the imbalance in the data. By default = None – no change in weights. balanced – automatically calculates weights based on class frequencies. This is useful for uneven distribution of classes.
- random_state – the starting point for generating random numbers (used for reproducibility of results).

```python
X = data.data[:, :2] 
y = data.target

h = 0.02 
color_dict = {0: "blue", 1: "lightgreen", 2: "red"}

colors = [color_dict[i] for i in y]
depth_2 = DecisionTreeClassifier(max_depth = 2).fit(X, y)
depth_4 = DecisionTreeClassifier(max_depth = 4).fit(X, y)
titles = ["Max depth = 2", "Max depth = 4"]
```

```python
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
```

```python
for i, classifier in enumerate((depth_2, depth_4)):
    plt.figure()
    Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap = plt.cm.coolwarm, alpha = 0.8)
    plt.scatter(X[:, 0], X[:, 1], c = colors)
    plt.xlabel("Sepal length")
    plt.ylabel("Sepal width")
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xticks(())
    plt.yticks(())
    plt.title(titles[i])
plt.show()
```

## Decision Forest 

### Voting Classifier
Voting Classifier combines the predictions of several classifiers. Each classifier is trained on all data and has different hyperparameters.

```python
trees = [
    DecisionTreeClassifier(max_depth=8),
    DecisionTreeClassifier(max_depth=4),
    DecisionTreeClassifier(max_depth=15),
    DecisionTreeClassifier(max_depth=10),
    DecisionTreeClassifier(max_depth=2),
]
```

```python
for tree in trees:
    tree.fit(digits_attributes, digits_target)
```

```python
for tree in trees:
    plot_tree(tree)
    plt.show()
```

```python
results = {}
for i, el in enumerate(trees):
    results[f"tree_{i}"] = el.predict(digits_attributes)
```

```python
df_results = pd.DataFrame(results)
```

```python
df_results
```

```python
# Majority vote for each record is class:
df_results["target"] = digits_target
```

```python
df_results
```

### Desicion Forest
Desicion Forest / Random Forest is an ensemble ML method that uses multiple Decision Trees that have similar hyperparameters. They are trained on different parts of the data and with different features. These parts are created by bootstrap sampling – at each tree part of the data is randomly selected with return.

```python
forest = RandomForestClassifier(n_estimators=7)
```

```python
forest.fit(digits_attributes, digits_target)
```

```python
for estimator in forest.estimators_:
    print(estimator.tree_.max_depth)
```

```python
# check how important each pixel is for the forest, sort by most importants (suitable for models that have a low bias):
pd.DataFrame(list(zip(data_digits.feature_names, np.round(forest.feature_importances_, 3)))).sort_values(by=1, ascending=False)
```

The more a pixel has participated in the splitting in the tree / forest the more important it is (high feature_importances_). This parameter has a problem if there are many unique values. For that reason its better to use **permutation_importance**. It is suitable for different models:

```python
p_i_mean = permutation_importance(forest, digits_attributes, digits_target)["importances_mean"]
```

```python
p_i_mean
```

```python
list(zip(data_digits.feature_names, np.round(p_i_mean, 3)))
```
