### FunctionTransformer

**FunctionTransformer()** - if I want to apply a specific transformation that is not covered by the standard transformers like StandardScaler, MinMaxScaler or OneHotEncoder, on some columns.

```python
log_transfomer = FunctionTransformer(np.log10)
```

```python
EPS = 1e-10
```

```python
# transformator for numerical columns, which i use in the main preprocessor:
number_processor = Pipeline([
    ("log_transfomer", FunctionTransformer(lambda x: np.log10(x + EPS))),
    ("min_max", MinMaxScaler((-5, 5)))
])
```

```python
number_processor
```

```python
preprocessor = ColumnTransformer([
    ("categorical", OneHotEncoder(), cat_columns),
    ("numerical", number_processor, num_columns)
])
```

```python
preprocessor
```

```python
pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("classifier", LogisticRegression())
])
```

## Model evaluation

The total error of a model consists of: Reducible error(Bias, Variance) and Irreducible error:

### Irreducible error
Irreducible error is unavoidable error – part of the total error of the model that cannot be reduced, no matter how complex or optimized the model is. Reasons for it are: Noise in the data (measurement errors); Lack of information (missing important variables that affect prediction); Complexity of the real world (there are phenomena that cannot be captured by even the most sophisticated mathematical models.

### Reducible error
- **Variance** (Statistical error) - variation in the data. Random error due to the fact that my data is diverse (external factors). It is a measure of how sensitive the model is to changes in the training data. High variance means the model knows the data so well that it becomes *chaotic* and performs poorly on new, unfamiliar data. It is obtained due to a too complex model - **overfitting** to the data. [Coeff. α (hypothesis testing)]
- **Bias** (Assumption error) – bias error due to a simplified model, bad data selection, bad learning process. The model fails to capture important dependencies in the data. This leads to poor performance of both training and test data (**underfitting**).

### Regularization
Regularization is a technique to prevent overfitting. It controls the complexity of the model by adding a 'penalty' (λ) to the loss function, which the model tries to minimize during its training. This causes the model to keep its parameters small and avoid large coefficients that would lead to excessive complexity. There are 2 types of regularization:
- L2 Regularization (RIDGE Regression) is the 2nd norm – adds the sum of the squares of the coefficients to the loss function. λ is a hyperparameter that controls the severity of the penalty (larger λ -> stronger penalty), and w are the coefficients of the model.

- L1 Regularization (LASSO Regression) is the 1st norm – adds the sum of the absolute values ​​of the coefficients to the loss function. It also reduces the complexity of the model, but it can also make some of the coefficients zero, which leads to feature selection. In this way, Lasso eliminates some variables from the model completely if they do not contribute significantly to the result.

- Elastic Net – a combination of Ridge and Lasso regression, adding both t(alpha = L2 and l1_ratio = L1)d L2) to the loss fun.ures.

```python
# L1 = 6 (= alpha * l1_ratio),  L2 = 1.5
# alpha = a + b  and  l1_ratio = a / (a + b)
ElasticNet(alpha = 1.5, l1_ratio = 4)
```

```python
# penalty=None - dont make regularization
LogisticRegression(penalty = None)
```

$ C = \frac{1}{λ} $ \
C is inverse of regularization strength. If it is a large number - there is almost no regularization (Ex: C=1e9 is almost the same as penalty=None). Big C = small λ. By default C=1 - it does regularization.

```python
iris = load_iris()
```

```python
attributes, labels = iris.data[:, [2, 3]], iris.target
weights, params = [], []
```

```python
for c in np.arange(-5, 5): 
    model = LogisticRegression(C=10.0 ** c)
    model.fit(attributes, labels)
    weights.append(model.coef_[1])  
    params.append(10.0 ** c) 

weights = np.array(weights)

plt.plot(params, weights[:, 0], label="Petal length")
plt.plot(params, weights[:, 1], label="Petal width")
plt.xlabel("C")
plt.ylabel("Weight coefficient")
plt.xscale("log")
plt.legend()
plt.show()
```

There are many metrics by which i can evaluate the **regression model**:
- R-squared (R² / Coefficient of Determination) – .score() shows how much of the variation in the target variable can be explained by the independent variables. The closer it is to 1, the better the model explains the data. A very high R² may indicate overfitting if not also verified on test data. Very low R² - the model does not explain the relationships in the data well.
- Mean Squared Error (MSE) – measures the mean squared difference between the model predictions and the actual values.
- Root Mean Squared Error (RMSE) is the square root of MSE and shows the errors in the same measurement unit as the dependent variable. The lower the values, the better the model predicts the data.
- Mean Absolute Error (MAE) – shows the mean absolute difference between the predicted and actual values. MAE is more robust to outliers than MSE and RMSE because it does not raise the squared errors.
- Adjusted R-squared – an adjusted variant of R² that takes into account the number of features in the model. If the addition of a new feature does not contribute to explaining the variation, this metric may decreres.
- Residual Standard Error (RSE) – the average error between forecasts and actual values. It shows how far the model predictions are from the true values ​​in the environtaset.
- Residual Plot – each point represents the difference between the actual values ​​and the model predictions (residuals). On the X-axis are the model predictions, on the Y-axis are the ress good.

There are many metrics by which i can evaluate the **classification model**:training.


- Accuracy: .score() gives the % of correctly classified examples (TP+TN) out of the total. High accuracy is desirable, but can sometimes be misleading for unbalanced classes - the model predicts the main class well but ignores rarer classes.
- Confusion Matrix – shows how the model classifies the data relative to the true classes. True Positives (TP) – number of correctly classified positive examples. True Negatives (TN) – number of correctly classified negative examples. False Positives (FP) – number of incorrectly classified positive examples (type I error). False Negatives (FN) – number of incorrectly classified negative examples (type II error).
- Precision – shows how many of the predicted positive (TP) examples are actually positive. With high precision – false positives (FP) are minimal.
- Recall (sensitivity) / TPR – shows how many of the actual positive examples (TP) were correctly detected by the model. High coverage means that the model was able to catch most positive examples.
- F1-Score – average value between precision and range (harmonic mean), balances the two metrics. It's useful when I need to find a balance between precision and recall; when I have imbalanced classes and want to avoid both many False Positives and False Negatives. F1-Score = $ 2 *  \frac{\text{Precision * Recall}}{\text{Precision + Recall}}$
- ROC Curve (Receiver Operating Characteristic) is a graph that shows the performance of a classification model by comparing True Positive Rate TPR (Recall) and False Positive Rate FPR at different probability thresholds. On the x-axis are FPRs. On the y-axis are TPRs. Each point on the curve represents a different probability threshold for how confident the model needs to be to decide whether an example is positive or negative. The dashed line on the diagonal shows a completely random model with AUC = 0.5 (as good as flipping a fair coin). The curve closer to the upper left corner indicates a better model. AUC = 1 perfect model that correctly predicts all positive and negative examples. AUC > 0.8 is good. An ROC curve gives the results at different thresholds, and the AUC summarizes how well the model performs across the range of thresholds. The higher the threshold (higher probability), the more certain the model is that a patient is sick, and accordingly will have smaller values ​​of TP and FP, but we may miss some truly sick people (more FN). If I set the threshold too high - I limit the model, it becomes too simple - bias (the model fails to sufficiently capture the complexity of the data). It will require a very high probability to classify someone as sick, which may result in many of the truly sick (FN) being missed. The model is biased towards predicting healthy patients.
- Log-Loss (Logarithmic Loss) – the model predicts the probability that an example belongs to the positive class (eg: instead of simply predicting "yes" / "no", the model can predict that the probability of "yes" is 0.8. More low Log-Loss – the model predicts probabilities close to the true classes.
- Cross-Validation (cross-validation) – gives a better idea of ​​how the model will work with new, unseen data; reduces the risk of overfitting. I divide all the data I have into 3 parts - the largest is for training (train), a smaller one (test) for tests and another small one, which I use only once at the very end of the process (validation / development set) to show how the model did. This is done if I have a lot of data. But if I have little data, I do the validation process several times (k-fold corss validation). Ex: data is divided into *k* subsets. The model is trained *k* times, each time one of the subsets is used for testing and the remaining *k* - 1 for training.

Choosing the best model - I try different settings, hyperparameters, optimizing them and choosing the most suitable for my case. I can automatically do this with:

GridSearchCV is a model hyperparameter optimization technique that automatically finds the best hyperparameters by creating a grid of all possible combinations of hyperparameters that I have provided in between. By default it uses cross-validation (KFold / StratifiedKFold) to estimate each combination of hyperparameters on the training set.
- estimator - the model I want to improve
- param_grid - hyperparameters which i want to optimize
- cv - number of folds
- scoring - the performance indicator by which different combinations of hyperparameters are evaluated. By default it uses .score(), but it can also use others (accuracy, precision, recall, f1, roc_auc)

```python
cv = GridSearchCV(estimator=pipeline, param_grid={
    "classifier__C": [1, 10, 100],
    "preprocessor__numerical__min_max__feature_range": [(-1, 1), (0, 1), (-5, 5)],
}, cv = 2).fit(sample_attrs2, sample_target2)
```
